### Pass by Object Reference

#### Formal and Actual Arguments
The actual arguments used in a function call are of 4 types:
 Positional arguments
 Keyword arguments
 Default arguments
 Variable length arguments

```python
# Positional arguments - These are the arguments passed to a function in correct positional order

def pos_example1 (s1,s2) :
    s3 = s1 + ' ' +  s2
    return s3

pos_example1('Santoshkumar', 'Pandit')

# change of arguments order affect

def pos_example2 (s2,s1) :
    s3 = s1 + ' ' +  s2
    return s3

pos_example2('Santoshkumar', 'Pandit')
```

```python
# Keyword Arguments - Keyword arguments are arguments that identify the parameters by their names. 

def grocery1(items,price):
    return items,price

grocery1(items = 'rice', price = 100)

# change of arguments order won't affect

def grocery2(items,price):
    return items,price

grocery2(price = 100, items = 'rice')
```

```python
# Default Arguments - We can mention some default values for the function parameters in the definition. 
# If pass in arguments, will take pass arguments value but if not passed than the default value in function will be used

def def_example (items, price = 100):
    return items, price

def_example(price = 200, items = 'Santosh')
```

```python
# Variable Length Arguments - . A variable length argument is an argument that can accept any number of values

def add(farg, *args):
    print('Formal arguments =' ,farg)
    sum=0
    for i in args:
        sum+=i
    print(sum)
    

add(10,20,30,40)
```

```python
# Local and Global Variables

# The Global Keyword - the global variable is not accessible inside the function but outside of it, it is accessible

# Namespaces and Scope in Python - We can picture a namespace as a Python dictionary structure, 
# where the dictionary keys represent the names and the dictionary values the object itself 

# The scope in Python defines the hierarchy level in which we search namespaces for certain name-to-object mappings
```

```python
#  if we declare a as global, all the reference and assignment go to the global a

def outer_function():
    global a
    a = 20
    
    def inner_function():
        global a
        a = 30
        print('a = ',a)
        
    inner_function()
    print('a = ',a)

global a # insite we declared here global a it took inner_function as global a since program execution understands that as global a
a = 10
outer_function()
print('a = ',a)
```

```python
# passing a group of elements to a fucntions

def passby_example (lst):
    sum1 = 0
    for i in lst:
        sum1 += i
    avg = sum1 / len(lst)
    return sum1, avg

print('Provide values in numbers till 10')
lst = list(int(x) for x in input().split( ))

x,y = passby_example(lst)
print('Total = ',x)
print('Average = ',y)
```

```python
# recursive functions 

def factorial (n):
    if n == 1:
        return 1
    else :
        y = n * factorial(n-1)
        return y

for x in range(1,10):
    print('Factorial of {} is {}'.format(x, factorial(x)))
```

```python
# filter expression - filter (function, sequence)

# Here, the function represents a function name that may return either True or False; and
# sequence' represents a list, string or tuple. The function' is applied to every element of the
# sequence' and when the function returns True, the element is extracted otherwise it is ignored


# Using Lambdas with filter() Function 

print('Provide values seperated by space')
lst1 = list(int(x) for x in input().split())
filter_example = list(filter(lambda x : (x%2==0),lst1))
print(filter_example)

# using user-defined function in filter

def is_even (x):
    if x%2 == 0:
        return True
    else:
        return False

print('Provide values seperated by space')
lst2 = list(int(x) for x in input().split())

filter_example_function = list(filter(is_even,lst2)) # here filter() needs first arguments and then iterables. The function itself does not need to be called with the list as an argument explicitly; filter() handles that by applying the function to each element of the list lst2.
print(filter_example_function)
```

```python
# Using Lambdas with map() Function - map(function, sequence)

# The function' performs a specified operation on all the elements of the sequence and the modified elements are returned which can be stored in another sequence

def squares1 (x):
    return x*x

lst_map = [1,2,3,4]
map_function = list(map(squares1,lst_map))
print(map_function)

# using lamda in map function

lst_maplambda = [1,3,4,5]
map_function1 = list(map(lambda x : x*x, lst_maplambda))
print(map_function1)

lst_maplambda1 = [1,3,4,5]
lst_maplambda2 = [2,6,7,8]
map_function1 = list(map(lambda x,y : x*y, lst_maplambda1,lst_maplambda2))
print(map_function1)
```

# Exception handling in Python - an exception is run-time error handled by programmer

```python
# Errors in a Python Program

# 1. Compile-time errors - detected by Python Compiler 
# 2. Runtime errors - detected by run-time error
# 3. Logical errors - 

# compile-time errors and logical errors can be solved by programmer but runtime errors which can be handled is called exception handling
```

```python
# all exceptions are represented as classes in python

# exceptions already in python - built-in exceptions and its base class is BaseException

# derivation :- (class)BaseException -----> (sub-class)Exception -----> (sub-class)StandardError & Warning

# all errors or exceptions are defined as sub-classes of StandardError and warnings as subclasses of Warning

# warnings can be neglected but errors cannot
```

```python
# below examples to remember three statements for exception handling in any programs

try:
    f = open('myfile.txt', 'w')
    a,b = [int(x) for x in input('Enter two number').split()]
    c = a/b
    print('The result is',c)
    f.write("The results is" + str(c))

except ZeroDivisionError :
    print("Error of division")
    print("Please re-execute the code programs")
    
finally :
    f.close()
    print('The file is closed')
```

# the complete exception handling syntax in below format :-


try:
    statements
    
except Exception1 :
    handler1
    
except Exception2 :
    handler2
    
else:
    statements
    
finally:
    statements
